```python
#import data to the hurricanes variable
import csv

#copied from https://automatetheboringstuff.com/2e/chapter16
def process_csv(filename):
    exampleFile = open(filename, encoding="utf-8")
    exampleReader = csv.reader(exampleFile)
    exampleData = list(exampleReader)
    exampleFile.close()
    return exampleData

hurricanes = process_csv('hurricanes.csv')
```

```python
header = hurricanes[0]
hurricanes = hurricanes[1:]
```

**Trick 1: How do I modify a list while looping?**

```python
#header
#Task: Make the items in the header allcaps
#Try 1
for item in header:
    item = item.upper()
print(header)
#Nothing changes!
```

```python
#Try 2
for item in header:
    header[item] = item.upper()
print(header)
#TypeError! item is a string, not an int
```

```python
#Try 3
for ind in range(len(header)): # <-- This is the key line!
    header[ind] = header[ind].upper()
print(header)
#This works!
```

**Trick 2: What happens if I pop() while looping?**

```python
#Task: Generate a list of only FAST hurricanes >=100 MPH
#Try 1
for index in range(len(hurricanes)):
    if int(hurricanes[index][3]) < 100:
        #Too slow! Delete
        hurricanes.pop(index)
print(hurricanes)
```

```python
#Try 2
hurricanes = process_csv('hurricanes.csv')[1:]
print("Length of hurricanes:",len(hurricanes))
for index in range(len(hurricanes)):
    if index >= len(hurricanes):
        break
    if int(hurricanes[index][3]) < 100:
        #Too slow! Delete
        hurricanes.pop(index)
print("New length of hurricanes:",len(hurricanes))
hurricanes
#Deleted some hurricanes, but not all that met the criteria
```

```python
#Try 2a
hurricanes = process_csv('hurricanes.csv')[1:]
print("Length of hurricanes:",len(hurricanes))
for cane in hurricanes:
    #if int(cane[3]) < 100:
    print(cane)
for index in range(len(hurricanes)):
    if index >= len(hurricanes):
        break
    if int(hurricanes[index][3]) < 100:
        #Too slow! Delete
        hurricanes.pop(index)
print("New length of hurricanes:",len(hurricanes))
for cane in hurricanes:
    if int(cane[3]) < 100:
        print(cane)
        
#What's happening: Whenever we delete, we advance to the next index
#Which means we don't check the element right after whatever we just deleted
```

```python
#Try 3
hurricanes = process_csv('hurricanes.csv')[1:]
for index in range(len(hurricanes)):
    if index >= len(hurricanes):
        break
    if int(hurricanes[index][3]) < 100:
        #Too slow! Delete
        hurricanes.pop(index)
        #WANT to instruct index to go back a step
        index -= 1 #Unfortunately, this doesn't work! (There's an advanced technique like this)
hurricanes
```

```python
#Try 4
#Approach: Instead of using pop, build a new list of the things we don't want to pop
hurricanes = process_csv('hurricanes.csv')[1:]
new_hurricanes = []
for cane in hurricanes:
    if int(cane[3]) >= 100:
        new_hurricanes.append(cane)
hurricanes = new_hurricanes
hurricanes
#This works!
#The secret: Do the modifying (deleting) on a fresh list that's not the one I'm looping over
```

**Trick 3: What happens if I modify a list passed as an argument?**

```python
head = process_csv('hurricanes.csv')[0]

def print_header_in_all_caps(header):
    for ind in range(len(header)): # <-- This is the key line!
        header[ind] = header[ind].upper()
    print(" ".join(header))
    
print_header_in_all_caps(head)
print(head) #head is also changed! Major side-effect!

#This is a property of MUTABLE types (lists)
```

Notes on passing lists: Either...
 - avoid modifying the list I get as an argument, or
 - make myself a new copy by copying all elements of the list
 - (be careful if you have a list of lists, the inner lists, also mutable, also can have this side-effect)
