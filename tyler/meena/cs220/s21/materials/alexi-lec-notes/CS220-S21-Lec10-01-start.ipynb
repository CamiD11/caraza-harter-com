```python
def format_day(day):
    #For the purpose of date conversaion this simple condition will work.
    if 10 <= day <= 20:
        return str(day) + "th"
    elif day % 10 == 1:
        return str(day) + "st"
    elif day % 10 == 2:
        return str(day) + "nd"
    elif day % 10 == 3:
        return str(day) + "rd"
    else:
        return str(day) + "th"
```
