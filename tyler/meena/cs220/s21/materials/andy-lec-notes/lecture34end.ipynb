```python
import sqlite3, os, pandas as pd

db_name = "bus.db"
assert os.path.exists(db_name)

conn = sqlite3.connect(db_name)

df = pd.read_sql("select * from sqlite_master", conn)

# note to self...we need to close this when we are done
```

```python
df
```

```python
# this command shows how our database was set up
# Note the table names, we will use them later
# Also note the column names
# finally, note that types are more formal, different than Python, and strictly enforced

for command in df["sql"]:
    print(command)
```

```python
# pandas shows its awesomeness again
# pandas allows us to write SQL query and create a DataFrame
pd.read_sql("select * from routes", conn)
```

```python
# repeat the step above but do this for boarding
pd.read_sql("select * from boarding", conn)
```

```python
pd.read_sql("select Route, DailyBoardings from boarding", conn)
```

```python
# more complicated queries

query = "select StopID, DailyBoardings from boarding where Route=14 order by DailyBoardings DESC limit 10"
pd.read_sql(query, conn)
```

```python
# queries can be written as their own strings
# multi-line strings in Python can be written using 3 "

query = """
    select StopID, DailyBoardings, Lat, Lon
    from boarding
    order by DailyBoardings DESC limit 10
"""
bus_riders = pd.read_sql(query, conn)
bus_riders.head()
```

```python
query = """
    select sum(DailyBoardings)
    from boarding
"""
total_bus_riders = pd.read_sql(query, conn)
total_bus_riders
```

```python
# west most bus route

query = """
    select *
    from boarding
    order by Lon ASC
"""
western_bus_routes = pd.read_sql(query, conn)
western_bus_routes
```
