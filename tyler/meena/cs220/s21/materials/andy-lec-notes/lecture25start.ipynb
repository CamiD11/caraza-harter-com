```python
# example 1: A generator is a function whose purpose is to generate a sequence of values

def f():
    yield 1    # yield is like return,
    yield 2    # but each time the function is called...
    yield 3    # it executes where it left off


print(f())     # the function cannot be called in the typical way


#for x in f():  # it has to be called through iteration (e.g. a for loop) 
    #print(x)

print("Yay, complete!")
```

```python
# example 2: Generators are what we call "Lazy" functions...they do just enough work
#.           and stop when they get to a 'yield'

def g():
    print("A")
    yield 1
    print("B")
    yield 2
    print("C")
    yield 3


for x in g():
    print(x)
    
print(g())

print("Yay, complete!")
```

```python
# example 3:  a generator is restarted when called again

def lets_go_red():
    yield "lets"
    yield "go"
    yield "red!"

for x in lets_go_red():
    print(x)

for x in lets_go_red():
    print(x)

# can you write a nested loop that prints lets...go...red!    n times?

n = 5
# your code here

```

```python
# example 4:  The 'next' function works on generators
# https://www.w3schools.com/python/ref_func_next.asp

g = lets_go_red()
print(g)
print(next(g))
print(next(g))
print(next(g))

# what happens if we copy/paste the last line  ?
```

```python
# example 5:  each generator keeps track of its own next value

g1 = lets_go_red()
g2 = lets_go_red()

print(next(g1))
print(next(g1))

print(next(g2))
print(next(g2))
```
