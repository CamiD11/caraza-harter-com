```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3]) # vectors (1D)
a
```

```python
len(a)
```

```python
a.shape
```

```python
b = np.array([[1,2,3], [4,5,6]]) # matrices (2D)
b
```

```python
len(b)
```

```python
b.shape
```

```python
b.reshape(3, 2)
```

```python
b.reshape(6)
```

```python
b.reshape(3, -1) # infer 2 columns
```

```python
b.reshape(-1)
```

```python
np.ones(80).reshape(10, -1)
```

```python
c = np.array(3) # scalar (0D)
c.shape
```

```python
# tensor: general term for any N dimensions
# images can be either 2D or 3D tensors
```

```python
a = np.ones(80).reshape(10, -1)
a
```

```python
plt.imshow(a, cmap="gray") # color map associates colors with numbers
```

```python
a = np.arange(80).reshape(8, -1)
a
```

```python
plt.imshow(a, cmap="gray") # color map associates colors with numbers
```

```python
a = plt.imread("bug.jpg")
plt.imshow(a)
```

```python
a.shape # each pixel is represented by a red/green/blue mix
```

```python
plt.imshow(a[400:, :2000, :])
```

```python
# MAKE THE RED PARTS BRIGHT
plt.imshow(a[:, :, 0], cmap="gray") # 0:red, 1:green, 2:blue
```

```python
# MAKE THE GREEN PARTS BRIGHT
plt.imshow(a[:, :, 1], cmap="gray") # 0:red, 1:green, 2:blue
```

```python
# MAKE THE BLUE PARTS BRIGHT
plt.imshow(a[:, :, 2], cmap="gray") # 0:red, 1:green, 2:blue
```
