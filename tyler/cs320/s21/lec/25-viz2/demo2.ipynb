# Shapefiles

```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
city = geopandas.read_file("zip://city.zip") # still same dir as my code (relative path)
lakes = geopandas.read_file("zip://lakes.zip")
police = geopandas.read_file("zip://police.zip")
fire = geopandas.read_file("zip://fire.zip")
```

```python
area = city.plot(color="0.85", figsize=(10,10))
lakes.plot(ax=area, color="lightblue")
fire.plot(ax=area, color="red", marker="+", label="Fire")
police.plot(ax=area, color="blue", label="Police")
area.set_axis_off()
plt.legend(loc="upper left", frameon=False)
fig = area.get_figure()
print(fig.dpi)
fig.savefig("high.png", dpi=300, bbox_inches="tight")
fig.savefig("low.png", dpi=20, bbox_inches="tight")
```
