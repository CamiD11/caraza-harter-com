```python
# pca
df = orig.copy().iloc[:, :-1]
df
```

```python
components = pd.DataFrame(np.random.randint(0, 100, (3, 5)), columns=[f"x{i}" for i in range(5)])
components
```

```python
from sklearn.decomposition import PCA
p = PCA(3)
weights = p.fit_transform(df)
pc = pd.DataFrame(weights, columns=["pc0", "pc1", "pc2"]).round(1)
pc.round().astype(int)
```

```python
comp = pd.DataFrame(p.components_, columns=df.columns).round(1)
comp
```

```python
pc.values @ comp.values
```

```python
(df - p.mean_).round().astype(int)
```
