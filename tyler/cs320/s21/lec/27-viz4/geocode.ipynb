```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv("events.csv")
df["Address"] = df["Address"].str.replace("Madison", "; Madison").str.replace("Block ", "")
df
```

```python
gpd.tools.geocode("1700 Parkside Drive; Madison, WI", provider='nominatim',
                  user_agent="cs320")
```

```python
geo = gpd.tools.geocode(df["Address"], provider='nominatim',
                        user_agent="cs320")
df["geometry"] = geo["geometry"]
df = gpd.GeoDataFrame(df)
```

```python
df[["Incident", "Date", "geometry"]].to_file("events.shp")
```
