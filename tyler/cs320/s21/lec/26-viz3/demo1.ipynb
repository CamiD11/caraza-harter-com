# Shapely

```python
import geopandas
from matplotlib import pyplot as plt
from shapely.geometry import Polygon, Point
from descartes import PolygonPatch
```

```python
Point(1, 1)
```

```python
a_set = {1,2,3}
b_set = {3,4,5,6,7,8}

qa = len(a_set.difference(b_set))
qb = len(b_set.difference(a_set))
qab = len(a_set.intersection(b_set))
qmax = max(qa, qb, qab)

fig, ax = plt.subplots()
ax.set_xlim(0,3)
ax.set_ylim(0,2)
c1 = Point(1, 1).buffer(1)
c2 = Point(2, 1).buffer(1)
AB = c1.intersection(c2)
A = c1.difference(c2)
B = c2.difference(c1)

ax.text(1, 2, "A", size=32, va="bottom", ha="center")
ax.text(2, 2, "B", size=32, va="bottom", ha="center")

for quantity, area in [(qa,A), (qb,B), (qab,AB)]:
    perc = (1 - quantity / qmax) * 0.6 + 0.4
    ax.add_artist(PolygonPatch(area, facecolor=str(perc)))
    ax.text(area.centroid.x, area.centroid.y, quantity, size=16)
plt.axis("off")
```
