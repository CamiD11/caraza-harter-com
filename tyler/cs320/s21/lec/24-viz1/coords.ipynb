```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
p = plt.Circle((0.5, 0.5), 0.1, edgecolor="red", linewidth=3, facecolor="none")
ax.add_artist(p) # a patch is a kind of artist
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8,4))
ax2.set_xlim(-0.1, 3)

def rel_circ(area, x, y):
    p = plt.Circle((x, y), 0.1, edgecolor="red",
                   linewidth=3, facecolor="none")
    area.add_artist(p) # a patch is a kind of artist
    
def abs_circ(area, x, y):
    p = plt.Circle((x, y), 0.2, facecolor="blue",
                   transform=fig.dpi_scale_trans)
    area.add_artist(p) # a patch is a kind of artist
    
rel_circ(ax1, 0.5, 0.5)
rel_circ(ax2, 0.5, 0.5)
rel_circ(fig, 0.5, 0.5)
abs_circ(fig, 4, 2)
x, y = ax2.transData.transform((2, 0.5)) / fig.dpi
abs_circ(fig, x, y)

abs_circ(fig, 1, 1)
abs_circ(ax1, 1, 2)
```
