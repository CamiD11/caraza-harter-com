```python
import os, time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
options = Options()
b = webdriver.Chrome(options=options)
```

```python
# https://stackoverflow.com/questions/43382447/python-with-selenium-drag-and-drop-from-file-system-to-webdriver

JS_DROP_FILE = """
    var target = arguments[0],
        offsetX = arguments[1],
        offsetY = arguments[2],
        document = target.ownerDocument || document,
        window = document.defaultView || window;

    var input = document.createElement('INPUT');
    input.type = 'file';
    input.onchange = function () {
      var rect = target.getBoundingClientRect(),
          x = rect.left + (offsetX || (rect.width >> 1)),
          y = rect.top + (offsetY || (rect.height >> 1)),
          dataTransfer = { files: this.files };

      ['dragenter', 'dragover', 'drop'].forEach(function (name) {
        var evt = document.createEvent('MouseEvent');
        evt.initMouseEvent(name, !0, !0, window, 0, 0, 0, x, y, !1, !1, !1, !1, 0, null);
        evt.dataTransfer = dataTransfer;
        target.dispatchEvent(evt);
      });

      setTimeout(function () { document.body.removeChild(input); }, 25);
    };
    document.body.appendChild(input);
    return input;
"""

def drag_and_drop_file(drop_target, path):
    driver = drop_target.parent
    file_input = driver.execute_script(JS_DROP_FILE, drop_target, 0, 0)
    file_input.send_keys(path)
```

```python
def upload(name):
    b.get("https://mediaspace.wisc.edu/upload/media")
    time.sleep(0.5)
    b.find_element_by_id("disclaimer-Accepted").click()
    time.sleep(0.5)
    drag_and_drop_file(b.find_element_by_id("uploadbox1"),
                       os.path.abspath(f"lec/{name}"))
    time.sleep(0.5)
    while b.find_element_by_id("cancelBtn").is_displayed():
        print(f"uploading {name}...")
        time.sleep(1)
    time.sleep(0.5)
    name_txt = b.find_element_by_id("Entry-name")
    name_txt.clear()
    name_txt.send_keys(f"cs220/{name}")
    name_txt.send_keys(Keys.RETURN)
    time.sleep(0.5)
    b.find_element_by_id("Entry-submit").click()
    time.sleep(3)
```

```python
videos = []
for root, _, names in os.walk("lec"):
    for name in names:
        if name.endswith(".mp4"):
            videos.append(f"{root}/{name}")
videos.sort()
print(len(videos))
videos[:5]
```

```python
t0 = time.time()
for i, video in enumerate(videos):
    upload(video[4:]) # slice off lec/
    t1 = time.time()
    print(f"{i+1} videos at {(t1-t0)/(i+1)} seconds per video")
```
